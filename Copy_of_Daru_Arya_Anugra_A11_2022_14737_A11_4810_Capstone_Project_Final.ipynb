9. Menangani Missing Values, Error, Duplikasi, dan Outlier

```python
# Tentukan kolom numerik dan kategorik
kolom_numerik = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
kolom_kategorik = [col for col in df.columns if col not in kolom_numerik + ['NObeyesdad']] # Anda sudah melakukan ini

# 1. Menangani Missing Values
# Anda sudah mengimputasi missing values di langkah sebelumnya,
# tetapi bisa diulang di sini untuk kejelasan atau jika dataframe di-load ulang.
print("Jumlah missing values sebelum imputasi:\n", df.isnull().sum())

# Imputasi median untuk numerik
for col in kolom_numerik:
    df[col] = df[col].fillna(df[col].median()) # Hindari inplace=True

# Imputasi modus untuk kategorik
for col in kolom_kategorikal: # Pastikan kolom_kategorikal sudah didefinisikan
    df[col] = df[col].fillna(df[col].mode()[0]) # Hindari inplace=True

print("\nJumlah missing values setelah imputasi:\n", df.isnull().sum())
assert df.isnull().sum().sum() == 0, "Masih ada missing value setelah imputasi!"


# 2. Menangani Error (Opsional - Tergantung dataset)
# Ini adalah bagian di mana Anda memeriksa nilai-nilai yang tidak logis.
# Contoh: Umur negatif, tinggi 0, dll.
# Jika dataset Anda bersih, Anda bisa melewati langkah ini atau menambahkannya jika Anda menemukan error.
# print("\nMemeriksa nilai-nilai yang tidak wajar...")
# Misalnya, cek umur di bawah umur minimal yang masuk akal
# if (df['Age'] < 1).any():
#     print("Peringatan: Ditemukan umur di bawah 1.")
# Anda bisa memutuskan untuk menghapus baris ini atau memperbaikinya.


# 3. Menangani Duplikasi
print("\nJumlah duplikasi sebelum penghapusan:", df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True) # Menghapus duplikat dan reset index
print("Jumlah duplikasi setelah penghapusan:", df.duplicated().sum())


# 4. Menangani Outlier
print("\nShape sebelum penghapusan outlier:", df.shape)
for col in kolom_numerik:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # Filter data untuk mempertahankan hanya nilai di dalam batas IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]

print("Shape setelah penghapusan outlier:", df.shape)

print("\n--- Proses Penanganan Data Selesai ---")
print("DataFrame Anda setelah preprocessing:")
print(df.head())
```

```python
# 5. Standarisasi data

# Pastikan Anda sudah menjalankan sel sebelumnya yang menghasilkan X_train dan X_test
from sklearn.preprocessing import StandardScaler # Pastikan ini sudah diimpor jika belum diimpor di sel ini

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- Tambahkan kode di bawah ini untuk melihat output ---
print("Shape data training setelah scaling:", X_train_scaled.shape)
print("Shape data testing setelah scaling:", X_test_scaled.shape)
# Opsional: Melihat beberapa nilai pertama (standarisasi menghasilkan array numpy)
# print("\n5 baris pertama data training setelah scaling:\n", X_train_scaled[:5])
# print("\n5 baris pertama data testing setelah scaling:\n", X_test_scaled[:5])
```

```python
# 6. Model sebelum tuning (baseline)
model_awal = RandomForestClassifier(random_state=42)
model_awal.fit(X_train_scaled, y_train)
y_pred_awal = model_awal.predict(X_test_scaled)
akurasi_awal = accuracy_score(y_test, y_pred_awal)

# --- Tambahkan kode di bawah ini untuk melihat output ---
print("Akurasi model sebelum tuning (baseline):", akurasi_awal)
```

```python
# 7. GridSearchCV - Hyperparameter Tuning
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train_scaled, y_train)
best_model = grid_search.best_estimator_
```

```python
# 8. Evaluasi model terbaik
# Pastikan variabel best_model, X_test_scaled, y_test, dan accuracy_score sudah tersedia dari sel-sel sebelumnya.

y_pred_best = best_model.predict(X_test_scaled)
akurasi_best = accuracy_score(y_test, y_pred_best)

# --- Tambahkan kode di bawah ini untuk melihat output ---
print("Akurasi model terbaik setelah tuning:", akurasi_best)
# Opsional: print beberapa prediksi dan nilai y_test sebenarnya
# print("\nContoh prediksi:", y_pred_best[:10])
# print("Contoh nilai sebenarnya:", y_test[:10].values)
```

```python
# 9. Visualisasi perbandingan
plt.figure(figsize=(7,5))
sns.barplot(x=["Sebelum Tuning", "Setelah Tuning"], y=[akurasi_awal, akurasi_best])
plt.ylim(0, 1)
plt.ylabel("Akurasi")
plt.title("Perbandingan Akurasi Sebelum dan Sesudah Hyperparameter Tuning")
plt.show()
```

```python

# 10. Classification Report
print("Akurasi Sebelum Tuning:", f"{akurasi_awal:.4f}")
print("Akurasi Setelah Tuning:", f"{akurasi_best:.4f}")
print("\nClassification Report Setelah Tuning:")
print(classification_report(y_test, y_pred_best))
```

```python
# 11. Kesimpulan
if akurasi_best > akurasi_awal:
    print(f"\n Tuning berhasil meningkatkan akurasi model dari {akurasi_awal:.2%} menjadi {akurasi_best:.2%}.")
    print("Model menjadi lebih optimal dalam memprediksi tingkat obesitas.")
else:
    print(f"\n Tuning tidak memberikan peningkatan signifikan. Akurasi tetap di sekitar {akurasi_best:.2%}.")
```
